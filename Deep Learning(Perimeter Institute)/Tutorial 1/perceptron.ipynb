## Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

## Parameters of the neuron

```python
N_in = 2

w = np.random.randn(N_in)
print(w.shape)
print(type(w))
print(w)
b = np.random.randn()
#print(b.shape)
print(type(b))
print(b)
```

## Define the neuron

```python
#Activation function ->>> SIGMOID
def f(z):
    return 1/(1+np.exp(-z))

def apply_neuron(x):
    z = np.dot(w,x) + b
    print(z.shape)
    print(z)
    return f(z)
```

## Sample the output of the neuron

```python
N_samples = 100 # Samples for each axis

samplex, sampley = np.meshgrid( 
    np.linspace(-10,10,N_samples), np.linspace(-10,10,N_samples)
)


sample_in = np.array([samplex.flatten(), sampley.flatten()])
print(type(sample_in))
print(sample_in.shape)
print(sample_in)
sample_out = apply_neuron(sample_in).reshape([N_samples,N_samples])
print(sample_out.shape)
print(sample_out)
```

## Visualize the results

```python
plt.figure()
plt.imshow(sample_out, extent=[-10,10,-10,10])
plt.xlabel("x")
plt.ylabel("y")
plt.show()

fig = plt.figure()
ax = fig.gca(projection="3d")
ax.plot_surface(samplex, sampley, sample_out ,rstride = 2, cstride = 2, color ="purple")
ax.plot_surface(samplex, sampley, np.zeros([N_samples,N_samples]),rstride = 2, cstride = 2, color ="blue")
ax.set_zlim(-0.1,1.1)
plt.show()
```
